```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# import the local python package
import os
import sys
sys.path.insert(0, os.path.abspath('..'))

import wild_bee_watch
```

```python
img = cv2.imread('../tests/data/single_bee.jpg')
```

```python
# Show the raw color image.
# Note: The image is captured with the Raspberry Pi NoIR camera which does not have an infrared filter. 
plt.imshow(img)
plt.show()
```

```python
# get the config that corresponds to the loaded image
config_loader = wild_bee_watch.config_loader.ConfigLoader()
config = config_loader.load('../tests/data/config.yaml')

# instantiate the preprocessing
preprocessing = wild_bee_watch.preprocessing.Preprocessing(config['preprocessing'])
```

```python
# run the preprocessing
img_normalized, img_color = preprocessing.process(img)
```

```python
# the color image is only warped but the pixel values are equivalent to those in the original image
plt.imshow(img_color)
plt.show()
```

```python
# The normalized image does not have these local illumination and contrast variations 
# Therefore, this it will be used for further processing.
plt.imshow(img_normalized, cmap='gray')
plt.show()
```
